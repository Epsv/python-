```python
#create a class student with attributes name and age. create two objects and print their details.
class student:
    def __init__(self,name,age):
        self.name=name
        self.age=age

a=student("shubham",70)
print(a.name)
print(a.age)
```

```python
#create a class car with attributes brand and price. write a method show_detail() to print car details . create two objects and call the method.
class car:
    def __init__(self,brand,price):
        self.brand=brand
        self.price=price

    def show_detail(self):
        print("brand:",self.brand)
        print("price:",self.price)

a=car("bmw",5000000)
b=car("audi",6000000)
a.show_detail()
b.show_detail()
```

```python
#Create a class Teacher with:
#• name
#• subject
#Use constructor to initialize values and display them using a method.
class teacher:
    def __init__(self,name,subject):
        self.name=name
        self.subject=subject

    def display(self):
        print("name:",self.name)
        print("subject:",self.subject)
a=teacher("shubham","math")
a.display()
```

```python
#create a parent class person with a method show() that prints "person details". create a child class student inherits person and create an object of student and call the method
class person:
    def show(self):
        print("person details")
class student(person):
    pass

a=student()
a.show()
```

```python
#create a class vehicle with a method move(). create a class car that inherits vehicle and call the inherited method
class vehicle:
    def move(self):
        print("vehicle is moving")
class car(vehicle):
    pass
a=car()
a.move()
```

```python
#create a parent class animal with a method eat()
#create a child class dog that has its own method bark()
#call both methods using a dog object
class animal:
    def eat(self):
        print("animal is eating")

class dog(animal):
    def bark(self):
        print("dog is barking")
a=dog()
a.eat()
a.bark()
```
