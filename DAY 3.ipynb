```python
#count frequency of words in a sentence using dictionary
sentence=input("enter the sentence-")
words=sentence.split()
freq = {}
for word in words:
    if word in freq:
        freq[word] +=1
    else:
        freq[word] = 1
print("word frequency:")
for word, count in freq.items():
    print(word, ":",count)
```

```python
#write a function that takes two numbers and returns their maximum
def max(a,b):
    if a>b:
        return a
    else:
        return b
max(2,6)
```

```python
#write a function that takes list of numbers and returns their average
def a(numbers):
    total = 0
    count = 0
    for num in numbers:
        total += num
        count += 1
    average = total / count
    return average
a([2,6,5,7,4])
```

```python
#write a function that counts vovels in a string
def v(string):
    count=0
    vovel="a,e,i,o,u,A,E,I,O,U"
    for char in string:
        if char in vovel:
            count += 1
    return count
v("hello")
```

```python
#write a function that calculates factorial of a number
def fac(a):
    b=a
    while a>1:
        b=b*(a-1)
        a-=1
    return b
fac(5)
```

```python
#write a function to check if a number is palindrome or not
def num(n):
    original=n
    reverse=0
    while n>0:
        digit=n%10
        reverse=reverse*10+digit
        n=n//10
    if reverse==original:
        print("it is palindrome number")
    else:
        print("it is not a  palindrome number")

    
num(65)
```
